```python
df = pd.read_csv('terrorismData.csv')
df.head()
city1 = df[(df['State']=="Jharkhand")|(df['State']=="Odisha" )|(df['State']=="Andhra Pradesh")|(df['State']=='Chhattisgarh')]    
city2 = df[df['State']=='Jammu and Kashmir']
casuality1 = (city1['Killed'] +city1['Wounded']).sum()
casuality2 = (city2['Killed'] +city2['Wounded']).sum()
year1 = set()
for i in df['Year']:
    year1.add(i)

length = len(year1)
print(int(casuality1/length),int(casuality2/length))
```
