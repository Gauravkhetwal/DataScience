```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,11,1)
y = np.arange(11,22,1)
plt.plot(x,y)
plt.grid
plt.figure()
plt.show
print(60,"to",70)
```

```python
x = np.arange(0,40,2)
y = 2 ** x
plt.plot(x,y ,"b--")
print(35)
```

```python
x = np.arange(0,6,1)
y1 = x**2
y2 = x**3

plt.plot(x,y1,color="red",label = "x2")
plt.plot(x,y2,color="blue",label = "x3")
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.axis([0,5,0,125])
plt.title("Matplotlib")
plt.grid()
plt.show()
```

### Microsoft bubble

```python
x = np.arange(0,10,1)
y = np.arange(20,30,1)
y2 = np.arange(0,40,2)
x2 = y2
plt.plot(x,y,color="blue",label = 'X')
plt.scatter(x2,y2,s=x2,color="grey",label = 'Y')
plt.axis([0,30,0,40])
plt.title("Matplotlib")
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.show()
```
