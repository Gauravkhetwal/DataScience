## Funding amount

```python
df = pd.read_csv('startup_funding.csv')

df.head(1)
```

```python
df.AmountInUSD.fillna(0, inplace = True)
df = df[df['CityLocation'].notna()]
df['CityLocation'].replace('Delhi','New Delhi',inplace = True)
df['CityLocation'].replace('bangalore','Bangalore',inplace = True)

def sepearteCity(city):
    return city.split('/')[0].strip()
df['CityLocation'] = df['CityLocation'].apply(sepearteCity)
df['AmountInUSD'] = df['AmountInUSD'].astype(str)
def removeComma(s):
    s = s.replace(',','')
    return s

df['AmountInUSD'] = df['AmountInUSD'].apply(removeComma)
df['AmountInUSD'] = df['AmountInUSD'].astype(float)
dict_Citywise_amount = {}
for i in range(len(df['CityLocation'])):
    dict_Citywise_amount[df['CityLocation'].iloc[i]] = dict_Citywise_amount.get(df['CityLocation'].iloc[i],0)+df['AmountInUSD'].iloc[i]
total_Amount_per_city = list(dict_Citywise_amount.values())
total_Amount_per_city.sort()

total_Amount_per_city = total_Amount_per_city[::-1]
total_Amount = sum(total_Amount_per_city[0:10])
total_Amount
def findKey(x):
    for i in list(dict_Citywise_amount.keys()):
        if(dict_Citywise_amount[i] == x):
            return i
for i in range(10):
    percentage_per_city = (total_Amount_per_city[i]/total_Amount)*100
    print(findKey(total_Amount_per_city[i]),format(percentage_per_city,'.2f'))

plt.pie(total_Amount_per_city[0:10])
plt.show()
```

## Investment Type

```python
df = pd.read_csv('startup_funding.csv')
df['InvestmentType'].replace('PrivateEquity','Private Equity',inplace = True)
df['InvestmentType'].replace('SeedFunding','Seed Funding',inplace = True)
df['InvestmentType'].replace('Crowd funding','Crowd Funding',inplace = True)
```

```python
df['InvestmentType'].describe()
```

```python
df.AmountInUSD.fillna(0,inplace = True)
df = df[df['InvestmentType'].notna()]
df['AmountInUSD'] = df['AmountInUSD'].astype(str)
df['AmountInUSD'] = df['AmountInUSD'].apply(removeComma)
df['AmountInUSD'] = df['AmountInUSD'].astype(float)
dict_investment = {}
for i in range(len(df['InvestmentType'])):
    dict_investment[df['InvestmentType'].iloc[i]]  = dict_investment.get(df['InvestmentType'].iloc[i],0)+ df['AmountInUSD'].iloc[i]

amount_per_investment = list(dict_investment.values())
total_Amount = sum(amount_per_investment)
def findKey(val):
    for i in dict_investment.keys():
        if dict_investment[i] == val:
            return i

for i in range(len(df['InvestmentType'].unique())):
    percenatage = (amount_per_investment[i]/total_Amount)*100
    print(findKey(amount_per_investment[i]),format(percenatage,'.2f'))
plt.pie(amount_per_investment,explode = [0.2,0.2,0.2,0.1],labels =df['InvestmentType'].unique() )
plt.show()
    
```
