## Number of fundings 

```python
df = pd.read_csv('startup_funding.csv')
df.head(2)
```

```python
df.AmountInUSD.fillna(0, inplace = True)
```

```python
df['Date']
```

```python
import csv

with open('startup_funding.csv','r') as file_obj:
    file_data = csv.DictReader(file_obj)
    file_list = list(file_data)

year = []
for row in file_list:
    year.append(row['Date'][-4::])
np_year = np.array(year,dtype = int)
df['Year'] = np_year
df.head(2)
```

```python
year = df['Year'].value_counts()
year_count = list(zip(year.index,year.values))
year_count = np.array(year_count)
year_count = year_count[year_count[:,0].argsort()]
required_year = year_count[:,0]
no_of_funding = year_count[:,1]
for i in range(len(required_year)):
    print(required_year[i],no_of_funding[i])
```

## plot the graph 

```python
plt.plot(required_year,no_of_funding,marker = 'o')
plt.xlabel("Year") ,
plt.ylabel("number of fundings")
plt.title("Year-wise Total Number of Fundings")
plt.xticks(rotation = 50)
plt.show()
```
